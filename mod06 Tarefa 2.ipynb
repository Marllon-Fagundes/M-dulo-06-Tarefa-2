```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

%matplotlib inline
```

```python
data = pd.read_csv('b1abb33c4f.csv')
```

```python
data.head()
```

```python
data.columns
```

####  1) Na base do SINASC do DataSUS, faça um boxplot da idade da mãe.

```python
fig, axes = plt.subplots(1, 1, figsize=(5, 5))
sns.boxplot(ax = axes, y='IDADEMAE', data=data)

axes.set_title('Idade da Mãe')
axes.set_ylabel('Idade')

plt.show()
```

####  2) Faça um boxplot do peso do bebê

```python
f, ax = plt.subplots(1, 1, figsize=(5, 5))
sns.boxplot(ax = ax, y='PESO', data=data)

ax.set_title('Peso do bebê')

plt.show()
```

####  3) Faça um histograma da idade da mãe

```python
f, ax = plt.subplots(1, 1, figsize=(5, 5))
sns.histplot(data=data, x='IDADEMAE' 
             , bins=20, kde=True
             , alpha = .25, element = 'step'
            )

ax.set_title('Idade das Mães')
ax.set_ylabel(None)
ax.set_xlabel('IDADE')

plt.show()
```

#### 4) Faça um histograma do peso do bebê

```python
f, ax = plt.subplots(1, 1, figsize=(5, 5))
sns.histplot(data = data, x='PESO')

ax.set_title('Peso dos bebês') 
ax.set_ylabel(None)

plt.show()
```
